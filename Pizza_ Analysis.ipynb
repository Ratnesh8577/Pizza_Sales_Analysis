## üî• HARD PIZZA SALES ANALYSIS

```python
# 1Ô∏è‚É£ Revenue Concentration (Pareto ‚Äì 80/20 Rule)
pizza_revenue = (
    df.groupby("pizza_name")["total_price"]
      .sum()
      .sort_values(ascending=False)
)

cumulative_pct = pizza_revenue.cumsum() / pizza_revenue.sum() * 100

pareto_df = pd.DataFrame({
    "revenue": pizza_revenue,
    "cumulative_percentage": cumulative_pct
})

pareto_df
```

```python
# 2Ô∏è‚É£ Price Elasticity Proxy (High Price vs Low Quantity)
price_quantity = (
    df.groupby("pizza_name")[["unit_price", "quantity"]]
      .mean()
      .sort_values("unit_price", ascending=False)
)

price_quantity
```

```python
# 3Ô∏è‚É£ Category Dependency Risk
category_share = (
    df.groupby("pizza_category")["total_price"]
      .sum()
)

dependency_ratio = category_share.max() / category_share.sum()
print("Revenue dependency ratio:", dependency_ratio)
```

```python
# 4Ô∏è‚É£ Menu Efficiency Score (Revenue per Ingredient)
ingredient_count = (
    df["pizza_ingredients"]
    .str.split(",")
    .apply(len)
)

df["ingredients_count"] = ingredient_count

menu_efficiency = (
    df.groupby("pizza_name")[["total_price", "ingredients_count"]]
      .sum()
)

menu_efficiency["revenue_per_ingredient"] = (
    menu_efficiency["total_price"] / menu_efficiency["ingredients_count"]
)

menu_efficiency.sort_values("revenue_per_ingredient", ascending=False)
```

```python
# 5Ô∏è‚É£ Sales Volatility (Demand Stability)
sales_volatility = (
    df.groupby("pizza_name")["quantity"]
      .std()
      .sort_values(ascending=False)
)

sales_volatility
```

```python
# 6Ô∏è‚É£ Peak Hour Revenue Density
hourly_density = (
    df.groupby("hour")
      .agg(
          total_revenue=("total_price", "sum"),
          total_orders=("order_id", "nunique")
      )
)

hourly_density["revenue_per_order"] = (
    hourly_density["total_revenue"] / hourly_density["total_orders"]
)

hourly_density.sort_values("revenue_per_order", ascending=False)
```

```python
# 7Ô∏è‚É£ Size‚ÄìCategory Profitability Matrix
size_category_matrix = pd.pivot_table(
    df,
    values="total_price",
    index="pizza_category",
    columns="pizza_size",
    aggfunc="sum",
    fill_value=0
)

size_category_matrix
```

```python
# 8Ô∏è‚É£ Cannibalization Detection
category_price_overlap = (
    df.groupby(["pizza_category", "pizza_size"])["unit_price"]
      .mean()
)

category_price_overlap
```

```python
# 9Ô∏è‚É£ Seasonal Strength Index
monthly_revenue = df.groupby("month")["total_price"].sum()
seasonal_index = monthly_revenue / monthly_revenue.mean()

seasonal_index
```
